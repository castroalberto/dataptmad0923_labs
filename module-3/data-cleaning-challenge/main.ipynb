```python
# COMPROBACION DE NULOS

for columna in iris_df.columns:
    nulos = iris_df[columna].isnull().sum()
    unicos = iris_df[columna].nunique()
    print(f"Columna {columna}: {nulos} nulos, {unicos} valores Ãºnicos")
```

```python
iris_df['class'].unique()
```

```python
correction_map = {
    'Iris-setossa': 'setosa',
    'Iris-setosa': 'setosa',
    'Iris-versicolor': 'versicolor',
    'versicolor': 'versicolor',
    'Iris-virginica': 'virginica'
}

iris_df['class'] = iris_df['class'].map(correction_map).fillna(iris_df['class'])

iris_df['class'].unique()
```

```python
iris_df.info()
```

```python
iris_df = iris_df.dropna()
```

```python
iris_df.info()
```

```python
iris_df
```

```python
iris_df['is_setosa'] = iris_df['class'] == 'setosa'
iris_df['is_virginica'] = iris_df['class'] == 'virginica'
iris_df['is_versicolor'] = iris_df['class'] == 'versicolor'
```

```python
iris_df.drop('is_versicolor', axis=1, inplace=True)
iris_df.drop('class', axis=1, inplace=True)
```

```python
iris_df
```

```python

features = ['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']

scaler = StandardScaler()

iris_df[features] = scaler.fit_transform(iris_df[features])
```
